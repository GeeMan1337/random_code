```python
import math as mt
import numpy as np
from matplotlib import pyplot as plt

c=299792458 #in metres per sec  #speed of light

#inputs
l=700 #in nanometres  #wavelength of light
D=1.5 #in metres  #distance to the screen
w=0.0001 #in metres #distance between slits
E=1 #amplitude of the electric wave
t=0.971 #time 
yrange=0.007 #in metres  #yrange on the screen 

y=np.linspace(-yrange,yrange,10000000)
o=2*mt.pi*c/l #in hertz #angular frequency omega
k=2*mt.pi/(l*10**(-9))  #in per metres  #2pi times wave number
z1=E*np.sin(k*(D*D+(y-w/2)**2)**0.5-o*t)  #wave from slit 1
z2=E*np.sin(k*(D*D+(y+w/2)**2)**0.5-o*t)  #wave from slit 2
I=(z1+z2)**2

plt.plot(y,I)
plt.show()
```
